```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters for the two assets
sigma_1 = 1  # Volatility of asset 1
mu_1 = 1.5   # Return of asset 1

sigma_2 = 2  # Volatility of asset 2
mu_2 = 2.5   # Return of asset 2

# Create the parameter s, ranging from -1 to 2 to cover short selling
s = np.linspace(-1, 2, 500)

# Case 1: rho_12 = -1
sigma_v_rho_neg1 = np.abs((1 - s) * sigma_1 - s * sigma_2)
mu_v_rho_neg1 = (1 - s) * mu_1 + s * mu_2

# Case 2: rho_12 = 1
sigma_v_rho_1 = np.abs((1 - s) * sigma_1 + s * sigma_2)
mu_v_rho_1 = (1 - s) * mu_1 + s * mu_2

# Create the plot
plt.figure(figsize=(10, 6))

# Plot rho_12 = -1
plt.plot(sigma_v_rho_neg1, mu_v_rho_neg1, label=r'$\rho_{12} = -1$', linestyle='--', color='blue')

# Plot rho_12 = 1
plt.plot(sigma_v_rho_1, mu_v_rho_1, label=r'$\rho_{12} = 1$', linestyle='--', color='green')

# Highlight the segment where 0 < s < 1 (no short selling) for both cases
s_no_short = np.linspace(0, 1, 500)
sigma_v_rho_neg1_no_short = np.abs((1 - s_no_short) * sigma_1 - s_no_short * sigma_2)
mu_v_rho_neg1_no_short = (1 - s_no_short) * mu_1 + s_no_short * mu_2
plt.plot(sigma_v_rho_neg1_no_short, mu_v_rho_neg1_no_short, color='blue', linewidth=2)

sigma_v_rho_1_no_short = np.abs((1 - s_no_short) * sigma_1 + s_no_short * sigma_2)
mu_v_rho_1_no_short = (1 - s_no_short) * mu_1 + s_no_short * mu_2
plt.plot(sigma_v_rho_1_no_short, mu_v_rho_1_no_short, color='green', linewidth=2)

# Mark the points for the individual assets
plt.scatter([sigma_1, sigma_2], [mu_1, mu_2], color='red', zorder=5)
plt.text(sigma_1, mu_1, r'$(\sigma_1, \mu_1)$', fontsize=12, verticalalignment='bottom')
plt.text(sigma_2, mu_2, r'$(\sigma_2, \mu_2)$', fontsize=12, verticalalignment='bottom')

# Labels and title
plt.xlabel(r'$\sigma_V$')
plt.ylabel(r'$\mu_V$')
plt.title('Typical portfolio lines with $\\rho_{12} = -1$ and $\\rho_{12} = 1$')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
