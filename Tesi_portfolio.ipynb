```python
import numpy as np
import pandas as pd

# Define the given probabilities for each scenario
probabilities = np.array([0.2, 0.5, 0.3])

# Expected returns for the two assets in each scenario
returns_asset_1 = np.array([0.10, 0.12, 0.08])
returns_asset_2 = np.array([0.15, 0.10, 0.05])

# Expected returns (mu) for each asset (weighted average return)
mu_1 = np.dot(probabilities, returns_asset_1)
mu_2 = np.dot(probabilities, returns_asset_2)

# Variance of returns for each asset
var_1 = np.dot(probabilities, (returns_asset_1 - mu_1) ** 2)
var_2 = np.dot(probabilities, (returns_asset_2 - mu_2) ** 2)

# Standard deviations (sigma) for each asset
sigma_1 = np.sqrt(var_1)
sigma_2 = np.sqrt(var_2)

# Covariance between the two assets
cov_12 = np.dot(probabilities, (returns_asset_1 - mu_1) * (returns_asset_2 - mu_2))

# Correlation coefficient between the two assets
rho_12 = cov_12 / (sigma_1 * sigma_2)

# Portfolio weights (w1 and w2)
w1 = 0.6  # 60% invested in Asset 1
w2 = 0.4  # 40% invested in Asset 2

# Portfolio expected return (mu_V)
mu_V = w1 * mu_1 + w2 * mu_2

# Portfolio variance (sigma_V^2)
sigma_V2 = w1**2 * var_1 + w2**2 * var_2 + 2 * w1 * w2 * cov_12

# Portfolio standard deviation (sigma_V)
sigma_V = np.sqrt(sigma_V2)

# Create a table for comparison
data = {
    'Asset': ['Asset 1', 'Asset 2', 'Portfolio'],
    'Expected Return (mu)': [mu_1, mu_2, mu_V],
    'Variance (Var)': [var_1, var_2, sigma_V2],
    'Standard Deviation (sigma)': [sigma_1, sigma_2, sigma_V],
    'Correlation Coefficient (rho_12)': ['', '', rho_12]
}

# Convert to DataFrame for better presentation
table = pd.DataFrame(data)

# Display the results
print(table)
```
