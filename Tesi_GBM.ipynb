```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the GBM model
S0 = 100  # Initial stock price
mu = 0.05  # Drift (annual return)
sigma = 0.2  # Volatility (annualized standard deviation)
T = 1  # Time horizon in years
N = 1000  # Number of time steps
dt = T / N  # Time step
np.random.seed(42)  # For reproducibility

# Initialize stock price array
S = np.zeros(N)
S[0] = S0

# Simulate the stock price using the GBM formula
for t in range(1, N):
    epsilon = np.random.normal(0, 1)  # Simulate epsilon from N(0,1)
    dS = mu * S[t-1] * dt + sigma * S[t-1] * epsilon * np.sqrt(dt)
    S[t] = S[t-1] + dS

# Plot the simulated stock price path
plt.figure(figsize=(10, 6))
plt.plot(np.linspace(0, T, N), S, label="Simulated Stock Price Path", color="b")
plt.title("Monte Carlo Simulation of Stock Price (GBM)")
plt.xlabel("Time (years)")
plt.ylabel("Stock Price")
plt.grid(True)
plt.legend()
plt.show()
```
