```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Constants
strike_price = 32000.00
current_index_value = 32018.82
index_step = 50
index_range = np.arange(31000, 33001, index_step)
price_per_index_point = 2.50
call_price = 1410.00
put_price = 885.00

# Function to calculate profit/loss for each position
def calculate_profit_loss(index_value, strike_price, call_price, put_price, price_per_index_point):
    intrinsic_value_call = max(0, index_value - strike_price) * price_per_index_point
    intrinsic_value_put = max(0, strike_price - index_value) * price_per_index_point

    # Long Call
    long_call_profit_loss = intrinsic_value_call - call_price

    # Short Call
    short_call_profit_loss = call_price - intrinsic_value_call

    # Long Put
    long_put_profit_loss = intrinsic_value_put - put_price

    # Short Put
    short_put_profit_loss = put_price - intrinsic_value_put

    return long_call_profit_loss, short_call_profit_loss, long_put_profit_loss, short_put_profit_loss

# List to store results
results = []

# Calculate profit/loss for each index value
for index_value in index_range:
    long_call_pl, short_call_pl, long_put_pl, short_put_pl = calculate_profit_loss(index_value, strike_price, call_price, put_price, price_per_index_point)
    results.append({
        "Index Value": index_value,
        "Long Call P/L": long_call_pl,
        "Short Call P/L": short_call_pl,
        "Long Put P/L": long_put_pl,
        "Short Put P/L": short_put_pl
    })

# Convert the list of results to a DataFrame
results_df = pd.DataFrame(results)

# Plotting the results for each position in separate figures
positions = ["Long Call", "Short Call", "Long Put", "Short Put"]
for position in positions:
    plt.figure(figsize=(12, 8))
    plt.plot(results_df["Index Value"], results_df[f"{position} P/L"], label=position)
    plt.xlabel("Index Value")
    plt.ylabel("Profit/Loss (â‚¬)")
    plt.title(f"Profit/Loss for {position} Position")
    plt.legend()
    plt.grid(True)
    plt.show()
```
